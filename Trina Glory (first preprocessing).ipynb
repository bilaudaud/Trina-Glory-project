```python
# Earliest date of 2016: '2016-01-02'
# Till date: k_pseudonymized['Erstellt am'].max() : '2021-04-11'
```

```python
# 1. filter from 2016 till date (lets get a count on data points)
k_pseudonymized = k_pseudonymized[(k_pseudonymized['Erstellt am'] >= '2016-01-01') & (k_pseudonymized['Erstellt am'] <= k_pseudonymized['Erstellt am'].max())]
```

```python
# 2. apply stratified sampling to divide the data into two sets 
# (1 set to be used for Expert Labels and the other to use medical rules label)

# let check the value counts of (age) column in other to stratify it across the division (Expert and medical rules) 
k_pseudonymized['age'].value_counts()
```

```python
k_pseudonymized_2016 = pd.read_csv('2016_k_pseudonymized.csv')
k_pseudonymized_2016.head()
```

```python
k_pseudonymized_2016['Erstellt am'] = pd.to_datetime(k_pseudonymized_2016['Erstellt am'])
k_pseudonymized_2016.info()
```

```python
# 2. apply stratified sampling to divide the data into two sets 
# (1 set to be used for Expert Labels and the other to use medical rules label)

# let check the value counts of (age) column in other to stratify it across the division (Expert and medical rules) 
k_pseudonymized_2016['age'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
expert_labels,medical_rules = train_test_split(k_pseudonymized_2016,test_size=0.5,stratify=k_pseudonymized_2016['age'])
```

```python
# percentage for the expert labelling
expert_labels['age'].value_counts()
```
