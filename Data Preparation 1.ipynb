```python
df1 = pd.read_csv('Data/k-pseudonymized-Draft1.csv',sep='\t',decimal=',')
df2 = pd.read_csv('Data/md_icd_pseudonymized_k5.csv',sep=',',decimal=',')

#df1.info(verbose=True)
```

```python
df1['Erstellt am'] = pd.to_datetime(df1['Erstellt am'])
```

```python
df_Join = df1.merge(df2, how='inner', left_index=True, right_on='index_exam')

df_Join.info(verbose=True)
```

```python
df_Join.head(5)
```

```python
mask = (df_Join['Erstellt am'] >= '2016-01-01')
```

```python
df_2016=df_Join.loc[mask]

df_2016.head(5)
```

```python
df_2016.info(verbose=True)
```

```python
#selecting important columns
columns = ['age','Weiblich/Männlich','Grübchenabstand DL-DR [mm]','Rumpflänge VP-DM [mm]','Rumpfneigung VP-DM [°]','Lotabweichung VP-DM [°]', 
'Seitabweichung VP-DM (RMS) [mm]','Oberflächenrotation (RMS) [°]','Kyphosewinkel VP-T12 [°]','Lordosewinkel T12-DM [°]',
'Beckenneigung (Symmetrielinie) [°]','Beckentorsion DL-DR [°]','Beckenhochstand [°]','Beckenrotation [°]','ICD']

df_2016_New = df_2016[columns]
```
