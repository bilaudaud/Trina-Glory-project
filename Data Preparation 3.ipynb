```python
import pandas as pd
import datetime as dt
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df1 = pd.read_csv('Data/k-pseudonymized-Draft1.csv',sep='\t',decimal=',')
df2 = pd.read_csv('Data/md_icd_pseudonymized_k5.csv',sep=',',decimal=',')

#df1.info(verbose=True)
```

```python
df1['Erstellt am'] = pd.to_datetime(df1['Erstellt am'])
```

```python
df1.info()
```

```python
df2.info()
```

```python
df_Join = df1.merge(df2, how='inner', left_index=True, right_on='index_exam')

df_Join.info()
```

```python
df_Join.head(5)
```

```python
mask = (df_Join['Erstellt am'] >= '2016-01-01')
```

```python
df_2016=df_Join.loc[mask]

df_2016.head(5)
```

```python
df_2016.info()
```

```python
#selecting important columns
columns = ['UUID','age','Weiblich/Männlich','Grübchenabstand DL-DR [mm]','Rumpflänge VP-DM [mm]','Rumpfneigung VP-DM [°]','Lotabweichung VP-DM [°]', 
'Seitabweichung VP-DM (RMS) [mm]','Oberflächenrotation (RMS) [°]','Kyphosewinkel VP-T12 [°]','Lordosewinkel T12-DM [°]',
'Beckenneigung (Symmetrielinie) [°]','Beckentorsion DL-DR [°]','Beckenhochstand [°]','Beckenrotation [°]','ICD','index_exam']

df_2016_New1 = df_2016[columns]
```

```python
df_2016_New1.head(5)
```

```python
df_2016_New1.isna().sum()
```

```python
df_2016_New1.fillna(df_2016_New1.median(), inplace=True)
```

```python
df_2016_New1.to_csv('Data/2016_Filtered-Draft1.csv', sep='\t', encoding='utf-8')
```

#### Due to the last data points which are unique; this has to be duplicated so that we have similar data in on X and Y set

```python
df_2016_dup = (df_2016_New1.query("ICD in ('G58.0','M48.00','M54.99','M47.84')"))

df_2016_dup.head()
```

```python
df_2016_New = pd.concat([df_2016_New1, df_2016_dup], ignore_index=True)
df_2016_New.shape

df_2016_New.to_csv('Data/2016_Filtered-Draft3.csv', sep='\t', encoding='utf-8')
#df_2016_New['ICD'].value_counts()
#df_2016_New['bc'] = df_2016_New['ICD'].astype(str) + "_" + df_2016_New['index_exam'].astype(str)

#df_2016_New['bc'].value_counts()

df_2016_New['ICD'].unique(), df_2016_New['ICD'].nunique()
```

```python
df_2016_New['ICD'].nunique()
```

```python
Labelled,UnLabelled = train_test_split(df_2016_New, random_state = 0, test_size = 0.75, stratify=df_2016_New['ICD'])

Labelled.shape,Labelled['ICD'].nunique()

#Unlabelled.head() 
```

```python
Labelled.drop_duplicates(subset=['ICD'], keep='last', inplace=True)

Labelled.to_csv('Data/ICDlabelled_Filtered-Draft1.csv', sep='\t', encoding='utf-8')

Labelled.shape,Labelled['ICD'].nunique()
```

```python
Labelled.drop_duplicates(subset=['index_exam'], keep='last', inplace=True)

Labelled.shape,Labelled['ICD'].nunique(),Labelled['index_exam'].nunique(),Labelled['UUID'].nunique(),Labelled['age'].nunique()
```

```python
Labelled.to_csv('Data/ICD_Labelled_Filtered-Draft2.csv', sep='\t', encoding='utf-8', index=False)
```
